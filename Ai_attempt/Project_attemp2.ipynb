```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import pythainlp
from pythainlp import word_tokenize, Tokenizer
from pythainlp.corpus.common import thai_stopwords
from pythainlp.ulmfit import document_vector
import re
```

```python
data = pd.read_excel('Abstract.xlsx')
print(data.columns)
```

```python
# abstract = ''
# for i in range(0, 10):
#   abstract += data.at[i,'บทคัดย่อ']
# print(abstract)

abstract = []
for i in range(0,19):
    sub = [data.at[i,'บทคัดย่อ']]
    abstract.append(sub)
```

```python
# print(abstract)
```

```python
def remove(text):
    text = text.replace("-",'')
    return re.sub(r"[\n|\t|\d|%|(|)|,|.|/|ฯ]{1,}|[A-Z]|[a-z]", "", text)

rm_abstract = []
for i in range(0,len(abstract)):
    sub = [remove(abstract[i][0])]
    rm_abstract.append(sub)
```

```python
# print(rm_abstract)
```

```python
token_abstract = []
for i in range(0,len(abstract)):
    sub = word_tokenize(rm_abstract[i][0], engine="deepcut", keep_whitespace=False)
    token_abstract.append(sub)
```

```python
# print(token_abstract)
```

```python
sw_abstract = []
sub = []
for i in range(0,len(abstract)):
    for j in range(0,len(token_abstract[i])):
        word = token_abstract[i][j]
        if word not in thai_stopwords():
            sub.append(word)
    sw_abstract.append(sub)
    sub = []
```

```python
print(sw_abstract)
```

## TF-IDF

1. tf(t,d) = number of occurrences of term in documnent / total number of all words in document 

```python
from collections import Counter
TFval = []
for i in range(0,len(abstract)):
    TFval.append({x:float(sw_abstract[i].count(x)/(len(sw_abstract[i]))) for x in sw_abstract[i]})
print(TFval)
```

2.idf(t,d) = log(how many times the term, t, appears / number of documents containind the term, t)

```python
IDF_term={}
for i in range(0,len(sw_abstract)):
    for j in range(0,len(sw_abstract[i])):
        if(sw_abstract[i][j] in IDF_term):
            IDF_term[sw_abstract[i][j]]+=int(1)
        else:
            IDF_term[sw_abstract[i][j]]=int(1)
# print(IDF_term)
```

```python
IDF_doc = {} 
for term in IDF_term:
    for i in range(0,len(sw_abstract)):
        if(term in sw_abstract[i]):
            if(term in IDF_doc):
                IDF_doc[term]+=int(1)
            else:
                IDF_doc[term]=int(1)
# print(IDF_doc)
#(DF)
```

```python
import math
IDFval = {}
for key in IDF_term.keys():
    IDFval[key] = math.log((IDF_term[key]+1)/(IDF_doc[key]+1),10)+1
# print(IDFval)
```

```python
TF_IDF = []
for i in range(0,len(abstract)):
    TF_IDF.append({key:TFval[i][key]*IDFval[key] for key in TFval[i]})
# print(TF_IDF)
```

```python
data = pd.DataFrame(columns=['Key','TF','IDF', "DF", 'TF_IDF'])
for i in range(0,len(abstract)):
    for key in TFval[i]:
        data = data.append({"Key":key, "TF":TFval[i][key], "IDF":IDFval[key], "DF":IDF_doc[key], "TF_IDF":TF_IDF[i][key]},ignore_index=True)
display(data)
#     data = pd.DataFrame(columns=['Key','TF','IDF', "DF",'TF_IDF'])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.xlabel('df(key)')
plt.ylabel('idf(key)')
plt.plot(data['DF'],data['IDF'],'o')
```
