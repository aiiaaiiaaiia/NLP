### Preprocessing
1. remove all punctuation   ( ) , : °   <br>
2. remove multiple newlines <br>
3. English part   <br>
Lowercase  <br>
4. Thai part   <br>
Normalize character order: for instance นำ้ to น้ำ

```python
import pythainlp
import re
import string
```

```python
#1 
def rm_punc(text):
    text = re.sub( r'\(', "", text)
    text = re.sub( r'\)', "", text)
    text = re.sub( r'\,', "", text)
    text = re.sub('\°','', text)
    text = re.sub('\:','',text)
    return text

# clean_text = [rm_punc(txt) for txt in text_list]
```

```python
#2
def rm_newline(text):
    return re.sub(r"[\n]{1,}", " ", text)
```

```python
#3 
def lowercase(text):
    return text.lower()
```

```python
#4
from pythainlp.util import normalize as normalize_char_order
# normalize_char_order('นำ้')   
```

```python
clean_text = [rm_punc(text) for text in text_list]
clean_text = [rm_newline(text) for text in clean_text]

# # join together
# blank = ''
# clean_text = blank.join(clean_text)
# print(clean_text)
```
